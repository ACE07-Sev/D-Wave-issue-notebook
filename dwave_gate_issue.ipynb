### Data class

Class that performs classical data preprocessing. Kindly ignore.

```python
from __future__ import annotations

__all__ = ['Data']

from abc import ABC
from collections.abc import Iterable
import numpy as np
from numpy.typing import NDArray
import math
from PIL import Image as Img
import matplotlib.pyplot as plt

# Data is either a vector or image
data_type = Iterable[float] | Iterable[Iterable[float]] | NDArray[np.number]


class Data(ABC):
    """ `Data` is the class used for accessing and manipulating data with
    built-in features such as normalization, and padding.
    """
    def __init__(self,
                 data: data_type) -> None:
        """ Initialize a `Data` instance.

        Parameters
        ----------
        `data` (data_type):
            The datapoint values.
        `norm_scale` (float):
            The normalization scale.
        `normalized` (bool):
            Whether the Data is normalized to 2 norm.
        `padded` (bool):
            Whether the Data is padded to a power of 2.
        `shape` (tuple[int]):
            The shape of the datapoint.
        """
        # Load in Data
        if not isinstance(data, np.ndarray):
            self.data = np.array(data)

        else:
            self.data = data

        # Save the data shape
        self.shape = self.data.shape

        # Set the norm scale (for normalization and denormalization)
        self.norm_scale = np.linalg.norm(self.data.flatten())

        # Set the normalized status
        self.is_normalized()

        # Set the padded status
        self.is_padded()

    @staticmethod
    def check_normalization(data: NDArray[np.number]) -> bool:
        """ Check if a data is normalized to 2-norm.

        Parameters
        ----------
        `data` (NDArray[np.number]):
            The data.

        Returns
        -------
        (bool): Whether the vector is normalized to 2-norm or not.
        """
        # Flatten the data in case it is a 2d array
        data_vector = data.flatten()

        # Check whether the data is normalized to 2-norm
        sum_check = np.sum(data_vector**2)

        # Check if the sum of squared of the data elements is equal to
        # 1 with 0.01 tolerance
        return math.isclose(sum_check, 1, abs_tol = 0.01)

    def is_normalized(self) -> None:
        """ Check if a `Data` object is normalized to 2-norm.
        """
        # If data is normalized, set `normalized` to True, Otherwise,
        # set `normalized` False
        self.normalized = self.check_normalization(self.data)

    @staticmethod
    def normalize_data(data: NDArray[np.number],
                       norm_scale: float,
                       data_shape: tuple[int]) -> NDArray[np.number]:
        """ Normalizes the data to 2-norm, and returns the
        normalized data.

        Parameters
        ----------
        `data` (NDArray[np.number]):
            The data.
        `norm_scale` (float):
            The normalization scale.
        `data_shape` (tuple[int]):
            The shape of the data.

        Returns
        -------
        (NDArray[np.number]): The 2-norm normalized data.
        """
        # Flatten the data in case it is a 2d array
        data_vector = data.flatten()

        # Normalize the vector to 2-norm, We then format to
        # return an np.array
        normalized_vector = [*(data_vector/norm_scale)]

        # Return the normalized data
        return np.array(normalized_vector).reshape(data_shape)

    def normalize(self) -> None:
        """ Normalizes a `Data` object to 2-norm.
        """
        # If the data is already normalized
        if self.normalized:
            return

        # If the data is not normalized, normalize the data
        self.data = self.normalize_data(self.data, self.norm_scale, self.shape)

        # Set normalized to True
        self.normalized = True

    @staticmethod
    def denormalize_data(data: NDArray[np.number],
                         norm_scale: float,
                         data_shape: tuple[int]) -> NDArray[np.number]:
        """ Denormalizes the data from 2-norm, and returns the
        denormalized data.

        Parameters
        ----------
        `data` (NDArray[np.number]):
            The 2-norm normalized datapoint.
        `norm_scale` (float):
            The normalization scale.
        `data_shape` (tuple[int]):
            The shape of the data.

        Returns
        -------
        (NDArray[np.number]): The denormalized data.
        """
        # Flatten the data in case it is a 2d array
        data_vector = data.flatten()

        # Denormalize the vector by applying the inverse of the
        # normalization factor
        denormalized_vector = [*(data_vector * norm_scale)]

        # Return the denormalized data
        return np.array(denormalized_vector).reshape(data_shape)

    def denormalize(self) -> None:
        """ Denormalizes a `Data` object from 2-norm.
        """
        # If the data is already denormalized, or simply not normalized
        if self.normalized is False:
            return

        # If the data is normalized, denormalize the data
        self.data = self.denormalize_data(self.data, self.norm_scale, self.shape)

        # Set normalized to False
        self.normalized = False

    @staticmethod
    def check_padding(data: NDArray[np.number]) -> bool:
        """ Check if a data is normalized to 2-norm.

        Parameters
        ----------
        `data` (NDArray[np.number]):
            The data.

        Returns
        -------
        (bool): Whether the vector is normalized to 2-norm or not.
        """
        # If the data is a vector
        if data.ndim == 1:
            # Check if the length of the vector is a power of 2
            return len(data) == 2 ** math.ceil(math.log2(len(data)))

        # If the data is a matrix
        elif data.ndim == 2:
            # Define the rows and columns size
            rows, cols = data.shape
            # Calculate the target size
            target_size = 2 ** math.ceil(math.log2(rows * cols))
            # Check if the number of matrix elements is a power of 2,
            # set `padded` to True
            return rows == math.ceil(target_size / cols)

        else:
            raise ValueError("Data must be a vector or a matrix.")

    def is_padded(self) -> None:
        """ Check if a `Data` object is padded to a power of 2.
        """
        self.padded = self.check_padding(self.data)

    @staticmethod
    def pad_data(data: NDArray[np.number]) -> tuple[NDArray[np.number],
                                                    tuple[int, int]]:
        """ Pads data with zeros up to the nearest power of 2.

        Parameters
        ----------
        `data` (NDArray[np.number]):
            The data to be padded.

        Returns
        -------
        `padded_data` (NDArray[np.number]): The padded data.
        `data_shape` (tuple[int]) : The updated shape.
        """
        # If the data is a vector
        if data.ndim == 1:
            # Calculate the target size
            target_size = 2 ** math.ceil(math.log2(len(data)))

            # Pad the vector with 0s
            padded_data = np.pad(data, (0, target_size - len(data)), mode='constant')

            # Update data shape
            updated_shape = padded_data.shape

            # Return all info
            return padded_data, updated_shape

        # If the data is a matrix
        elif data.ndim == 2:
            # Define the rows and columns size
            rows, cols = data.shape

            # Calculate the target size
            target_size = 2 ** math.ceil(math.log2(rows * cols))

            # Calculate the number of rows and columns needed for the
            # padded matrix
            target_rows = target_size / cols
            target_cols = target_size / rows

            # Ensure the target rows and columns are integers (i.e., we
            # can't have 1.5 rows)
            if target_rows != int(target_rows) or target_cols != int(target_cols):
                # If the data is wider than tall
                if cols > rows:
                    # Calculate the number of rows needed for the
                    # padded matrix
                    rows = 2 ** math.ceil(math.log2(rows))

                    # Calculate the number of columns needed for the
                    # padded matrix (prioritize wider images)
                    cols = int(target_size/rows)

                else:
                    # Calculate the number of columns needed for the
                    # padded matrix
                    cols = 2 ** math.ceil(math.log2(cols))

                    # Calculate the number of rows needed for the
                    # padded matrix (prioritize taller images)
                    rows = int(target_size/cols)

                # If we cannot pad, we stretch the whole matrix to the
                # nearest power of 2
                resized_data = np.asarray(Img.fromarray(data).resize((rows, cols), 0))

                # Return all info
                return resized_data, resized_data.shape

            # Ensure the target rows and columns are int type
            target_rows = int(target_rows)
            target_cols = int(target_cols)

            # Pad the matrix with 0s
            if target_rows * cols == target_size:
                padded_data = np.zeros((target_rows, cols), dtype=data.dtype)
            else:
                padded_data = np.zeros((rows, target_cols), dtype=data.dtype)

            # Copy the original matrix into the top-left corner of the
            # padded matrix
            padded_data[:rows, :cols] = data

            # Update data shape
            updated_shape = padded_data.shape

            # Return all info
            return padded_data, updated_shape

        else:
            raise ValueError("Data must be a vector or a matrix.")

    def pad(self):
        """ Pads a `Data` object.
        """
        # Check if the data is already padded or not
        if self.padded:
            return

        # Pad the data
        self.data, self.shape = self.pad_data(self.data)

        # Set padded to True
        self.padded = True

    def draw(self) -> None:
        """ Draws the `Data` object.
        """
        plt.imshow(self.data)
        plt.show()

    def __repr__(self) -> str:
        """ Returns a string representation of the `Data` object.

        Returns
        -------
        (str): The string representation of the `Data` object.
        """
        return f'Data(data={self.data})'

    def __eq__(self, other_data: data_type | Data) -> bool:
        """ Checks if two `Data` objects are equal.

        Parameters
        ----------
        `other_data` (data_type | Data):
            The other data to compare with.

        Returns
        -------
        (bool): Whether the two `Data` objects are equal or not.
        """
        if not isinstance(other_data, Data):
            other_data = Data(other_data)

        return np.all(self.data == other_data.data)

    def __len__(self) -> int:
        """ Returns the length of the `Data` object.

        Returns
        -------
        (int): The length of the `Data` object.
        """
        return len(self.data.flatten())

    def __mul__(self,
                multiplier: float) -> None:
        """ Multiplies a `Data` object by a scalar.

        Parameters
        ----------
        `multiplier` (float):
            The scalar to multiply by.
        """
        # Multiply the data by the multiplier
        self.data = np.multiply(self.data, multiplier)
```

### D-Wave circuit wrapper

The class that will be using the `dwave.gate` package.

```python
from __future__ import annotations

__all__ = ['Circuit']

from abc import ABC, abstractmethod
from collections.abc import Iterable
from types import NotImplementedType
import numpy as np
from numpy.typing import NDArray


class Circuit(ABC):
    """ `Circuit` is the class for creating and manipulating gate-based circuits.
        This class is defined for external Quantum Circuit (QC) Frameworks.
    """
    def __init__(self,
                 num_qubits: int,
                 num_clbits: int) -> None:
        """ Initialize a `Circuit` instance.

        Parameters
        ----------
        `num_qubits` (int):
            Number of qubits in the circuit.
        `num_clbits` (int):
            Number of classical bits in the circuit.
        """
        # Define the number of quantum bits
        self.num_qubits = num_qubits
        # Define the number of classical bits
        self.num_clbits = num_clbits
        # Define the circuit
        self.circuit = None
        # Define the measurement status
        self.measured = False
        # Define the circuit log (list[dict])
        self.circuit_log = []

    @abstractmethod
    def X(self,
          qubit_indices: int | Iterable[int]) -> None:
        """ Apply a Pauli-X gate to the circuit.

        Parameters
        ----------
        `qubit_indices` (int | Iterable[int]):
            The index of the qubit(s) to apply the gate to.
        """
        pass

    @abstractmethod
    def H(self,
          qubit_indices: int | Iterable[int]) -> None:
        """ Apply a Hadamard gate to the circuit.

        Parameters
        ----------
        `qubit_indices` (int | Iterable[int]):
            The index of the qubit(s) to apply the gate to.
        """
        pass

    @abstractmethod
    def RY(self,
           angle: float,
           qubit_index: int) -> None:
        """ Apply a RY gate to the circuit.

        Parameters
        ----------
        `angle` (float):
            The rotation angle in radians.
        `qubit_index` (int):
            The index of the qubit to apply the gate to.
        """
        pass

    @abstractmethod
    def RZ(self,
           angle: float,
           qubit_index: int) -> None:
        """ Apply a RZ gate to the circuit.

        Parameters
        ----------
        `angle` (float):
            The rotation angle in radians.
        `qubit_index` (int):
            The index of the qubit to apply the gate to.
        """
        pass

    @abstractmethod
    def CX(self,
           control_index: int,
           target_index: int) -> None:
        """ Apply a Controlled Pauli-X gate to the circuit.

        Parameters
        ----------
        `control_index` (int):
            The index of the control qubit.
        `target_index` (int):
            The index of the target qubit.
        """
        pass

    @abstractmethod
    def MCRY(self,
             angle: float,
             control_indices: int | Iterable[int],
             target_indices: int | Iterable[int]) -> None:
        """ Apply a Multi-controlled RY gate to the circuit.

        Parameters
        ----------
        `angle` (float):
            The rotation angle in radians.
        `control_indices` (int | Iterable[int]):
            The index of the control qubit(s).
        `target_indices` (int | Iterable[int]):
            The index of the target qubit(s).
        """
        pass

    def vertical_reverse(self) -> None:
        """ Reverse the circuit vertically.
        """
        # Iterate over every operation, and change the index accordingly
        for operation in self.circuit_log:
            keys = ['target_indices', 'control_indices', 'qubit_indices', 'qubit_index', 'control_index', 'target_index']
            for key in keys:
                if key in operation:
                    if isinstance(operation[key], Iterable):
                        operation[key] = [(self.num_qubits - 1 - index) for index in operation[key]]
                    else:
                        operation[key] = (self.num_qubits - 1 - operation[key])

        # Update the circuit
        self.circuit = self.convert(type(self)).circuit

    def horizontal_reverse(self,
                           adjoint: bool=True) -> None:
        """ Reverse the circuit horizontally.

        Parameters
        ----------
        `adjoint` (bool):
            Whether or not to apply the adjoint of the circuit.
        """
        # Reverse the order of the operations
        self.circuit_log = self.circuit_log[::-1]

        # If adjoint is True, then multiply the angles by -1
        if adjoint:
            for operation in self.circuit_log:
                if 'angle' in operation:
                    operation['angle'] = -operation['angle']
                elif 'angles' in operation:
                    operation['angles'] = [-angle for angle in operation['angles']]

        # Update the circuit
        self.circuit = self.convert(type(self)).circuit

    def add(self,
            circuit: Circuit,
            qubit_indices: int | Iterable[int]) -> None:
        """ Append two circuits together in a sequence.

        Parameters
        ----------
        `circuit` (Circuit):
            The circuit to append to the current circuit.
        `qubit_indices` (int | Iterable[int]):
            The indices of the qubits to add the circuit to.
        """
        # The number of qubits must match the number of qubits in the circuit.
        assert len(qubit_indices) == circuit.num_qubits, "The number of qubits must match the number of qubits in the circuit."

        # Update the qubit indices
        for operation in circuit.circuit_log:
            keys = ['target_indices', 'control_indices', 'qubit_indices', 'qubit_index', 'control_index', 'target_index']
            for key in keys:
                if key in operation:
                    if isinstance(operation[key], Iterable):
                        operation[key] = [(qubit_indices[index]) for index in operation[key]]
                    else:
                        operation[key] = (qubit_indices[operation[key]])

        # Add the other circuit's log to the circuit log
        self.circuit_log.extend(circuit.circuit_log)

        # Create the updated circuit
        self.circuit = self.convert(type(self)).circuit

    @abstractmethod
    def measure(self,
                qubit_indices: int | Iterable[int]) -> None:
        """ Measure the qubits in the circuit.

        Parameters
        ----------
        `qubit_indices` (int | Iterable[int]):
            The indices of the qubits to measure.
        """
        pass

    @abstractmethod
    def run(self,
            **kwargs) -> dict:
        """ Run the circuit.

        Parameters
        ----------
        `kwargs` (dict):
            The keyword arguments for the run.

        Returns
        -------
        `result` (dict): The result of the circuit.
        """
        pass

    @abstractmethod
    def get_statevector(self) -> Iterable[float]:
        """ Get the statevector of the circuit.

        Returns
        -------
        `statevector` (Iterable[float]): The statevector of the circuit.
        """
        pass

    @abstractmethod
    def get_counts(self,
                   **kwargs) -> dict:
        """ Get the counts of the circuit.

        Parameters
        ----------
        `kwargs` (dict):
            The keyword arguments for the run.

        Returns
        -------
        `counts` (dict): The counts of the circuit.
        """
        pass

    @abstractmethod
    def draw(self) -> None:
        """ Draw the circuit.
        """
        pass

    @abstractmethod
    def get_depth(self) -> int:
        """ Get the depth of the circuit.

        Returns
        -------
        `depth` (int): The depth of the circuit.
        """
        pass

    def convert(self,
                circuit_framework: Circuit) -> Circuit:
        """ Convert the circuit to another circuit framework.

        Parameters
        ----------
        `circuit_framework` (Circuit):
            The circuit framework to convert to.

        Returns
        -------
        `converted_circuit` (Circuit): The converted circuit.
        """
        # Define the new circuit using the provided framework
        converted_circuit = circuit_framework(self.num_qubits, self.num_clbits)

        # Define a mapping between Qiskit gate names and corresponding methods in the target framework
        gate_mapping = {
            'X': converted_circuit.X,
            'H': converted_circuit.H,
            'RY': converted_circuit.RY,
            'RZ': converted_circuit.RZ,
            'CX': converted_circuit.CX,
            'MCRY': converted_circuit.MCRY,
            'measure': converted_circuit.measure
        }

        # Iterate over the gate log and apply corresponding gates in the new framework
        for gate_info in self.circuit_log:
            # Find gate name
            gate_name = gate_info['gate']

            # Slide dict to keep kwargs only
            gate_info = dict(list(gate_info.items())[1:])

            # Use the gate mapping to apply the corresponding gate
            gate_mapping[gate_name](**gate_info)

        # Return the converted circuit
        return converted_circuit
```

```python
from __future__ import annotations

__all__ = ['DwaveCircuit']

from collections.abc import Iterable
from typing import Optional
import numpy as np
import math
import matplotlib.pyplot as plt

# D-Wave imports
from dwave.gate import Circuit as DWCircuit
from dwave.gate.operations import *
from dwave.gate.simulator import simulate


class DwaveCircuit(Circuit):
    """ `DwaveCircuit` is the wrapper for using D-Wave's Gate-based framework.
    """
    def __init__(self,
                 num_qubits: int,
                 num_clbits: int) -> None:
        """ Initialize a `DwaveCircuit` instance.

        Parameters
        ----------
        `num_qubits` (int):
            Number of qubits in the circuit.
        `num_clbits` (int):
            Number of classical bits in the circuit.
        """
        # Define the number of quantum bits
        self.num_qubits = num_qubits
        # Define the number of classical bits
        self.num_clbits = num_clbits

        # Define the circuit
        self.circuit = DWCircuit(self.num_qubits, self.num_clbits)

        # Define the measurement status
        self.measured = False

        # Define the circuit log (list[dict])
        self.circuit_log = []

    def X(self,
          qubit_indices: int | Iterable[int]) -> None:
        """ Apply a Pauli-X gate to the circuit.

        Parameters
        ----------
        `qubit_indices` (int | Iterable[int]):
            The index of the qubit(s) to apply the gate to.
        """
        # Create a Pauli X gate
        x = X()

        # Check if the qubit_indices is a list
        if isinstance(qubit_indices, Iterable):
            # If it is, apply the X gate to each qubit in the list
            for index in qubit_indices:
                with self.circuit.context as (q, c):
                    x(q[index])
                # dwave.gate requires calling the unlock method to add more gates
                self.circuit.unlock()
        else:
            # If it's not a list, apply the X gate to the single qubit
            with self.circuit.context as (q, c):
                x(q[qubit_indices])
            # dwave.gate requires calling the unlock method to add more gates
            self.circuit.unlock()

        # Add the gate to the log
        self.circuit_log.append({'gate': 'X', 'qubit_indices': qubit_indices})
    
    def H(self,
          qubit_indices: int | Iterable[int]) -> None:
        """ Apply a Hadamard gate to the circuit.

        Parameters
        ----------
        `qubit_indices` (int | Iterable[int]):
            The index of the qubit(s) to apply the gate to.
        """
        # Create a Hadamard gate
        h = Hadamard()

        # Check if the qubit_indices is a list
        if isinstance(qubit_indices, Iterable):
            # If it is, apply the H gate to each qubit in the list
            for index in qubit_indices:
                with self.circuit.context as (q, c):
                    h(q[index])
                # dwave.gate requires calling the unlock method to add more gates
                self.circuit.unlock()
        else:
            # If it's not an list, apply the H gate to the single qubit
            with self.circuit.context as (q, c):
                h(q[qubit_indices])
            # dwave.gate requires calling the unlock method to add more gates
            self.circuit.unlock()

        # Add the gate to the log
        self.circuit_log.append({'gate': 'H', 'qubit_indices': qubit_indices})
    
    def RY(self,
           angle: float,
           qubit_index: int) -> None:
        """ Apply a RY gate to the circuit.

        Parameters
        ----------
        `angle` (float):
            The rotation angle in radians.
        `qubit_index` (int):
            The index of the qubit to apply the gate to.
        """
        # If the angle is zero or a multiple of 2 pi, do not apply the RY gate
        if angle == 0 or math.isclose(angle % (2 * np.pi), 0):
            return

        # Create an RY gate with the specified angle
        ry = RY(angle)
        # Apply the RY gate to the circuit at the specified qubit
        with self.circuit.context as (q, c):
            ry(q[qubit_index])

        # dwave.gate requires calling the unlock method to add more gates
        self.circuit.unlock()

        # Add the gate to the log
        self.circuit_log.append({'gate': 'RY', 'angle': angle, 'qubit_index': qubit_index})

    def RZ(self,
           angle: float,
           qubit_index: int) -> None:
        """ Apply a RZ gate to the circuit.

        Parameters
        ----------
        `angle` (float):
            The rotation angle in radians.
        `qubit_index` (int):
            The index of the qubit to apply the gate to.
        """
        # If the angle is zero or a multiple of 2 pi, do not apply the RZ gate
        if angle == 0 or math.isclose(angle % (2 * np.pi), 0):
            return

        # Create an RZ gate with the specified angle
        rz = RZ(angle)
        # Apply the RZ gate to the circuit at the specified qubit
        with self.circuit.context as (q, c):
            rz(q[qubit_index])

        # dwave.gate requires calling the unlock method to add more gates
        self.circuit.unlock()

        # Add the gate to the log
        self.circuit_log.append({'gate': 'RZ', 'angle': angle, 'qubit_index': qubit_index})

    def CX(self,
           control_index: int,
           target_index: int) -> None:
        """ Apply a CX gate to the circuit.

        Parameters
        ----------
        `control_index` (int):
            The index of the control qubit.
        `target_index` (int):
            The index of the target qubit.
        """
        # Create a Controlled-X gate
        cx = CX()
        # Apply the CX gate to the circuit at the specified control and target qubits
        with self.circuit.context as (q, c):
                cx(q[control_index], q[target_index])

        # dwave.gate requires calling the unlock method to add more gates
        self.circuit.unlock()

        # Add the gate to the log
        self.circuit_log.append({'gate': 'CX', 'control_index': control_index, 'target_index': target_index})

    def MCRY(self,
             angle: float,
             control_indices: int | Iterable[int],
             target_indices: int | Iterable[int]) -> None:
        """ Apply a MCRY gate to the circuit.

        Parameters
        ----------
        `angle (float):
            The rotation angle in radians.
        `control_indices` (int | Iterable[int]):
            The index of the control qubit(s).
        `target_indices` (int | Iterable[int]):
            The index of the target qubit(s).
        """
        # If the angle is zero or a multiple of 2 pi, do not apply the MCRY gate
        if angle == 0 or math.isclose(angle % (2 * np.pi), 0):
            return

        # Ensure control_indices and target_indices are always treated as lists
        control_indices = [control_indices] if isinstance(control_indices, int) else control_indices
        target_indices = [target_indices] if isinstance(target_indices, int) else target_indices

        # As of now, dwave.gate does not support MCRY gates. We can use the ControlledOperation class
        # to create a custom MCRY gate
        class MCRYControlledOp(ControlledOperation):
            _num_control = len(control_indices)
            _num_target = 1
            _target_operation = RY(angle)

        # Apply the MCRY gate to the circuit at the control and target qubits
        for i in range(len(target_indices)):
            with self.circuit.context as (q, c):
                MCRYControlledOp(control=([q[index] for index in control_indices]), target=q[i])

            # dwave.gate requires calling the unlock method to add more gates
            self.circuit.unlock()

        # Add the gate to the log
        self.circuit_log.append({'gate': 'MCRY', 'angle': angle, 'control_indices': control_indices, 'target_indices': target_indices})

    def measure(self,
                qubit_indices: int | Iterable[int]) -> None:
        """ Measure qubits in the circuit.

        Parameters
        ----------
        `qubit_indices` (int | Iterable[int]):
            The indices of the qubits to measure.
        """
        # Measure the qubits
        with self.circuit.context as (q, c):
            meas = Measurement(q) | c

        # dwave.gate requires calling the unlock method to add more gates
        self.circuit.unlock()

        # Add the measurement to the class
        self.measurement = meas

        # Add the measurement to the log
        self.circuit_log.append({'gate': 'measure', 'qubit_indices': qubit_indices})

    def run(self,
            num_shots: int,
            backend: Optional[object],
            seed: Optional[int]) -> dict:
        """ Run the circuit on the backend.

        Parameters
        ----------
        `num_shots` (int):
            The number of shots to run.
        `backend` (Optional[object]):
            The backend to run the circuit on. NOT USED.
        `seed` (Optional[int]):
            The seed for the simulator. NOT USED.

        Returns
        -------
        `result` (dict): The result of the circuit.
        """
        if self.measured is False:
            self.measure(range(self.num_qubits))

        # D-wave uses MSB convention for qubits, so we need to reverse the qubit indices
        self.vertical_reverse()

        # Remove the vertical reverse from log as it does not belong to circuit definition
        for operation in self.circuit_log:
            keys = ['target_indices', 'control_indices', 'qubit_indices', 'qubit_index', 'control_index', 'target_index']
            for key in keys:
                if key in operation:
                    if isinstance(operation[key], Iterable):
                        operation[key] = [(self.num_qubits - 1 - index) for index in operation[key]]
                    else:
                        operation[key] = (self.num_qubits - 1 - operation[key])

        # Run the circuit
        simulate(self.circuit)

        # Execute the circuit on the backend
        result = self.measurement.sample(range(self.num_qubits), num_shots)

        # Return the result
        return result

    def get_statevector(self) -> Iterable[float]:
        """ Get the state vector of the circuit.

        Returns
        -------
        `statevector` (Iterable[float]): The state vector of the circuit.
        """
        # D-wave uses MSB convention for qubits, so we need to reverse the qubit indices
        self.vertical_reverse()

        # Remove the vertical reverse from log as it does not belong to circuit definition
        for operation in self.circuit_log:
            keys = ['target_indices', 'control_indices', 'qubit_indices', 'qubit_index', 'control_index', 'target_index']
            for key in keys:
                if key in operation:
                    if isinstance(operation[key], Iterable):
                        operation[key] = [(self.num_qubits - 1 - index) for index in operation[key]]
                    else:
                        operation[key] = (self.num_qubits - 1 - operation[key])

        # Run the circuit
        simulate(self.circuit)

        # Define the state vector
        state_vector = np.real(self.circuit.state)

        # Return the state vector
        return state_vector

    def get_counts(self,
                   num_shots: int,
                   backend: Optional[object],
                   seed: Optional[int] = 0) -> dict:
        """ Get the counts of the circuit.

        Parameters
        ----------
        `num_shots` (int):
            The number of shots to run.
        `backend` (Optional[object]):
            The backend to run the circuit on. NOT USED.
        `seed` (Optional[int]):
            The seed for the simulator. NOT USED.

        Returns
        -------
        `counts` (dict): The counts of the circuit.
        """
        # Run the circuit
        counts = self.run(num_shots=num_shots, backend=backend, seed=seed)

        # Return the counts
        return counts

    def draw(self) -> plt.figure:
        """ Draw the circuit.
        """
        # `dwave.gate` does not support drawing circuits. We can use the `qiskit` library to draw the circuit.
        # Convert the circuit to qasm format
        qasm = self.circuit.to_qasm()

        # Draw the circuit
        qc = QuantumCircuit.from_qasm_str(qasm)

        # Return the circuit
        return qc.draw(output='mpl')

    def get_depth(self) -> int:
        """ Get the depth of the circuit.

        Returns
        -------
        (int): The depth of the circuit.
        """
        return len(self.circuit.circuit)
```

```python
""" Helper functions for the Shende encoder
"""
def rotations_to_disentangle(local_param: list[float]) -> tuple:
    """ Takes a list of local parameters and returns a list of global parameters.

    Parameters
    ----------
    `local_param` (list[float]):
        The list of local parameters.

    Returns
    -------
    (tuple): The tuple of global parameters.
    """
    remaining_vector = []
    thetas = []
    phis = []
    param_len = len(local_param)

    def bloch_angles(pair_of_complex: list[complex]) -> tuple:
        """ Takes a pair of complex numbers and returns the corresponding Bloch angles.

        Parameters
        ----------
        `pair_of_complex` (list[complex]):
            The list of complex numbers.

        Returns
        -------
        (tuple): The list of Bloch angles.
        """
        [a_complex, b_complex] = pair_of_complex
        a_complex = complex(a_complex)
        b_complex = complex(b_complex)
        mag_a = abs(a_complex)
        final_r = np.sqrt(mag_a**2 + np.absolute(b_complex) ** 2)

        if final_r < 1e-10:
            theta, phi, final_r, final_t = 0, 0, 0, 0

        else:
            theta = 2 * np.arccos(mag_a / final_r)
            a_arg = np.angle(a_complex)
            b_arg = np.angle(b_complex)
            final_t = a_arg + b_arg
            phi = b_arg - a_arg

        return final_r * np.exp(1.0j * final_t / 2), theta, phi

    for i in range(param_len // 2):
        (remains, add_theta, add_phi) = bloch_angles(local_param[2 * i : 2 * (i + 1)])
        remaining_vector.append(remains)
        thetas.append(-add_theta)
        phis.append(-add_phi)

    return remaining_vector, thetas, phis
```

```python
from abc import ABC, abstractmethod
from typing import Optional
import time
import numpy as np
import math


class Encoder(ABC):
    """ `Encoder` is the class used in QIMP's framework for encoding and decoding images.
    """
    def __init__(self,
                 circuit_framework: Circuit) -> None:
        """ Initalize a Encoder instance.

        Parameters
        ----------
        `circuit_framework` (Circuit):
            The quantum circuit framework.
        """
        # Define the QC framework
        self.circuit_framework = circuit_framework

        # Define the encoder protocol
        self.get_encoder()

    @abstractmethod
    def get_encoder(self) -> None:
        """ Define the `encoder`.
        """
        pass

    @abstractmethod
    def encode_state(self,
                     data: data_type) -> Circuit:
        """ Encode a classical data into a quantum state.

        Paramaters
        ----------
        `data` (data_type):
            The data to be encoded.

        Returns
        -------
        `circuit` (Circuit): The encoded state's circuit.
        """
        pass

    @abstractmethod
    def decode_state(self,
                     data: data_type,
                     encoded_state: Circuit,
                     log=True) -> Data:
        """ Decode a quantum state into the original classical data.

        Paramaters
        ----------
        `data` (data_type):
            The original data for denormalization purposes.
        `encoded_state` (Circuit):
            The encoded state's circuit.
        `log` (bool):
            Whether to print the log or not.

        Returns
        -------
        `decoded_state` (Data): The decoded data.
        """
        pass

    def __call__(self,
                 data: data_type) -> Circuit:
        """ Calls the instance of the class with the given data
        and returns the encoded state.

        Parameters
        ----------
        `data` (data_type):
            The data to be encoded.

        Returns
        -------
        (Circuit): The encoded state.
        """
        return self.encode_state(data)
```

```python
class Shende(Encoder):
    """ `Shende` is the `Encoder` subclass used for encoding data using
        Shende et al. (2004).
        ref : https://arxiv.org/abs/quant-ph/0406176.
        This approach requires log2(N) qubits, and has a O(N) scalability.
    """
    def get_encoder(self) -> None:
        def encode_shende(data: data_type) -> Circuit:
            """ Encode data into a quantum state using Shende approach.

            Paramaters
            ----------
            `data` (data_type):
                The data to be encoded.

            Returns
            -------
            `circuit` (Circuit): The encoded state's circuit.
            """
            # Define a QIMP.Data instance
            if not isinstance(data, Data):
                data = Data(data)

            # Normalize data to 2-norm (if required)
            if data.normalized is False:
                data.normalize()

            # Pad the data (if required)
            if data.padded is False:
                data.pad()

            # Flatten the data in case it is not 1-dimensional
            data = data.data.flatten()

            # Define the number of qubits (n qubits, where n = Log2(N), where N is the dimension of the vector)
            num_qubits = int(np.ceil(np.log2(len(data))))

            # Construct Shende circuit
            circuit = self.circuit_framework(num_qubits, num_qubits)

            def multiplex_ry(list_of_angles: list[float],
                             last_cnot=True) -> Circuit:
                """ Takes a list of angles and returns a circuit that applies the corresponding multiplexed Ry gate.

                Parameters
                ----------
                `list_of_angles` (list[float]):
                    The list of angles.
                `last_cnot` (bool):
                    Whether to apply the last CNOT gate or not.

                Returns
                -------
                `circuit` (Circuit): The multiplexed Ry gate's circuit.
                """
                # Calculate the number of angles
                num_angles = len(list_of_angles)

                # Define the number of qubits for the local state
                local_num_qubits = int(math.log2(num_angles)) + 1

                # Define the Multiplex circuit
                circuit = self.circuit_framework(local_num_qubits, local_num_qubits)

                # Define LSB and MSB
                lsb, msb = 0, local_num_qubits - 1

                # If there is only one qubit, we apply the target gate to that single qubit
                if local_num_qubits == 1:
                    circuit.RY(list_of_angles[0], 0)
                    return circuit

                # If there are two qubits, we apply the target gate to the first qubit and then apply a CNOT gate
                angle_weight = np.kron([[0.5, 0.5], [0.5, -0.5]], np.identity(2 ** (local_num_qubits - 2)))

                list_of_angles = angle_weight.dot(np.array(list_of_angles)).tolist()

                # Define the first half multiplexed RY circuit
                multiplex_1 = multiplex_ry(list_of_angles[0 : (num_angles // 2)], False)
                circuit.add(multiplex_1, range(local_num_qubits-1))
                circuit.CX(msb, lsb)

                # Define the second half multiplexed RY circuit
                multiplex_2 = multiplex_ry(list_of_angles[(num_angles // 2) :], False)

                if num_angles > 1:
                    multiplex_2.horizontal_reverse(adjoint=False)
                    circuit.add(multiplex_2, range(local_num_qubits-1))
                else:
                    circuit.add(multiplex_2, range(local_num_qubits-1))

                # If the last CNOT gate is required, we apply it
                if last_cnot:
                    circuit.CX(msb, lsb)

                return circuit

            def multiplex_rz(list_of_angles: list[float],
                             last_cnot=True) -> Circuit:
                """ Takes a list of angles and returns a circuit that applies the corresponding multiplexed RZ gate.

                Parameters
                ----------
                `list_of_angles` (list[float]):
                    The list of angles.
                `last_cnot` (bool):
                    Whether to apply the last CNOT gate or not.

                Returns
                -------
                `circuit` (Circuit): The multiplexed Ry gate's circuit.
                """
                # Calculate the number of angles
                num_angles = len(list_of_angles)

                # Define the number of qubits for the local state
                local_num_qubits = int(math.log2(num_angles)) + 1

                # Define the Multiplex circuit
                circuit = self.circuit_framework(local_num_qubits, local_num_qubits)

                # Define LSB and MSB
                lsb = 0
                msb = local_num_qubits - 1

                # If there is only one qubit, we apply the target gate to that single qubit
                if local_num_qubits == 1:
                    circuit.RZ(list_of_angles[0], 0)
                    return circuit

                # If there are two qubits, we apply the target gate to the first qubit and then apply a CNOT gate
                angle_weight = np.kron([[0.5, 0.5], [0.5, -0.5]], np.identity(2 ** (local_num_qubits - 2)))

                list_of_angles = angle_weight.dot(np.array(list_of_angles)).tolist()

                # Define the first half multiplexed RZ gate
                multiplex_1 = multiplex_rz(list_of_angles[0 : (num_angles // 2)], False)
                circuit.add(multiplex_1, range(local_num_qubits-1))
                circuit.CX(msb, lsb)

                # Define the second half multiplexed RZ gate
                multiplex_2 = multiplex_rz(list_of_angles[(num_angles // 2) :], False)

                # If there are more than one angle, apply a horizontal reverse (no adjoint) before adding the multiplex
                if num_angles > 1:
                    multiplex_2.horizontal_reverse(adjoint=False)
                    circuit.add(multiplex_2, range(local_num_qubits-1))
                else:
                    circuit.add(multiplex_2, range(local_num_qubits-1))

                # If the last CNOT gate is required, we apply it
                if last_cnot:
                    circuit.CX(msb, lsb)

                return circuit

            def gates_to_uncompute(params: list[float],
                                   num_qubits: int) -> Circuit:
                """ Takes a list of parameters and returns a circuit that applies the corresponding gates

                Parameters
                ----------
                `params` (list[float]):
                    The list of parameters.
                `num_qubits` (int):
                    The number of qubits.

                Returns
                -------
                `circuit` (Circuit): The circuit that applies the corresponding gates.
                """
                # Define the circuit
                circuit = self.circuit_framework(num_qubits, num_qubits)

                # Define the remaining parameters
                remaining_param = params

                # Iterate over all qubits
                for i in range(num_qubits):
                    # Calculate the parameters for disentangling
                    (remaining_param, thetas, phis) = rotations_to_disentangle(remaining_param)
                    # Set the last cnot to true
                    add_last_cnot = True

                    # Check if the norm of phis and thetas is not 0
                    if np.linalg.norm(phis) != 0 and np.linalg.norm(thetas) != 0:
                        # Set the last cnot to false
                        add_last_cnot = False

                    # Check if the norm of phis is not 0
                    if np.linalg.norm(phis) != 0:
                        # Apply the multiplexed RZ gate
                        rz_mult = multiplex_rz(phis, last_cnot=add_last_cnot)
                        circuit.add(rz_mult, range(i, num_qubits))

                    # Check if the norm of thetas is not 0
                    if np.linalg.norm(thetas) != 0:
                        # Apply the multiplexed RY gate
                        ry_mult = multiplex_ry(thetas, last_cnot=add_last_cnot)
                        # Apply a horizontal reverse (not adjoint)
                        ry_mult.horizontal_reverse(adjoint=False)
                        circuit.add(ry_mult, range(i, num_qubits))

                return circuit

            # Define the disentangling circuit
            disentangling_circuit = gates_to_uncompute(data, num_qubits)
            # Apply a horizontal reverse (adjoint)
            disentangling_circuit.horizontal_reverse()
            # Add the disentangling circuit to the initial circuit
            circuit.add(disentangling_circuit, range(num_qubits))

            return circuit

        self.encoder = encode_shende

    def encode_state(self,
                     data: data_type) -> Circuit:
        """ Encode a classical data into a quantum state.

        Paramaters
        ----------
        `data` (data_type):
            The data to be encoded.

        Returns
        -------
        `circuit` (Circuit): The encoded state's circuit.
        """
        return self.encoder(data)

    def decode_state(self,
                     data: data_type,
                     encoded_state: Circuit,
                     log=True) -> Data:
        """ Decode a quantum state into the original classical data.

        Parameters
        ----------
        `data` (data_type):
            The original data for denormalization purposes.
        `encoded_state` (Circuit):
            The encoded state's circuit.
        `log` (bool):
            Whether to print the log or not.

        Returns
        -------
        `decoded_state` (Data): The decoded data.
        """
        # Define a QIMP.Data instance
        if not isinstance(data, Data):
            data = Data(data)

        # Pad the data (if required)
        if data.padded is False:
            data.pad()

        if log:
            print(f"The depth of the circuit is : {encoded_state.get_depth()}")

        # Start timer (Measuring the circuit)
        start = time.time()

        # Run circuit
        state_vector = encoded_state.get_statevector()

        # Convert state vector to a Data instance to apply further processing
        state_vector = Data(state_vector)

        # End timer (Measuring the circuit)
        end = time.time()

        if log:
            print('Measurement done...')
            print(f'Time taken to measure : {end-start}')

        # Start timer (Decoding the state vector)
        start = time.time()

        # Denormalize the state vector and reshape the decoded state into the original shape
        decoded_state = Data.denormalize_data(state_vector.data, data.norm_scale, data.shape)

        # End timer (Decoding the state vector)
        end = time.time()

        if log:
            print('Decoding done...')
            print(f'Time taken to decode : {end-start}')

        # Return the decoded data
        return Data(decoded_state)
```

```python
class FRQI(Encoder):
    """ `FRQI` is the `Encoder` subclass used for encoding data using
        Flexible Representation for Quantum Images approach.
        ref: https://doi.org/10.1007/s11128-010-0177-y.
        This approach requires log2(N) + 1 qubits, and has a O(N^2) scalability.
    """
    def get_encoder(self) -> None:
        """ Uses frqi(data) to encode a data.
        """
        def encode_frqi(data: data_type) -> Circuit:
            """ Encode data into a quantum state using Flexible Representation of Quantum Images (N^2 scalability).

            Parameters
            ----------
            `data` (data_type):
                The data to be encoded.

            Returns
            -------
            `circuit` (Circuit): The encoded state's circuit.
            """
            # Define a Data instance
            if not isinstance(data, Data):
                data = Data(data)

            # Normalize data to 2-norm (if required)
            if data.normalized is False:
                data.normalize()

            # Pad the data (if required)
            if data.padded is False:
                data.pad()

            # Flatten the data in case it is not 1-dimensional
            data = data.data.flatten()

            # Define the number of qubits (n + 1 qubits, where n = Log2(N), where N is the dimension of the vector,
            # first n qubits encode the position, and the last qubit encodes the pixel value)
            num_qubits = int(np.ceil(np.log2(len(data)))) + 1

            # Convert pixel values to rotation angles
            angles = np.arcsin(data)

            # Define the quantum circuit
            circuit = self.circuit_framework(num_qubits, num_qubits)

            def flag(state: bin,
                     new_state: bin) -> int | None:
                """ Flags which qubits need to be applied the X-gate to change the state of the pixels positions
                qubits to the new state.

                Parameters
                ----------
                `state` (bin):
                    The binary string of current state.
                `new_state` (bin):
                    The binary string of the updated state.

                Returns
                -------
                `c` (int): The integer format of the binary string.
                """
                # Define the length of the binary string
                n = len(state)
                # Create an empty array
                c = np.array([])
                # Start to iterate n times
                for i in range(n):
                    # Check if the position i in the binary string is different
                    if state[i] != new_state[i]:
                        # if the position is different, we append the position to the array
                        c = np.append(c, int(i))
                # Return the integer format of the binary string
                if len(c) > 0:
                    return c.astype(int)
                else:
                    return c

            def flip(circuit: Circuit,
                     state: bin,
                     new_state: bin) -> None:
                """ Applies the X-gates to the corresponding qubits given by the `flag` function.

                Parameters
                ----------
                `circuit` (Circuit):
                    The quantum circuit.
                `state` (bin):
                    The binary string of current state.
                `new_state` (bin):
                    The binary string of the updated state.
                """
                # Flag the state
                c = flag(state, new_state)
                # If the array is not empty, we apply the X-gates to flagged qubits
                if len(c) > 0:
                    circuit.X(np.abs(c-(circuit.num_qubits-2)))

            # Define the binary string format
            bin_string = '{0:0' + str(num_qubits-1) + 'b}'

            # Specify the control qubits (which are the position qubits)
            control_indices = [i for i in range(num_qubits-1)]

            # Specify the target qubit (which is the pixel value qubit)
            target = control_indices[-1] + 1

            # Apply a Hadamard Layer (only to the position qubits)
            circuit.H(control_indices)

            # Encode the pixel angle values
            for i in range(len(angles)):
                # If it is the first iteration, we don't need to flip the qubits
                if i == 0:
                    # Apply the C^n(Ry) (controlled Ry gate)
                    circuit.MCRY(2*angles[i], control_indices, target)
                else:
                    # Flip the qubits
                    state = bin_string.format(i-1)
                    new_state = bin_string.format(i)
                    flip(circuit, state, new_state)
                    # Apply the C^n(Ry) (controlled Ry gate)
                    circuit.MCRY(2*angles[i], control_indices, target)

            # Return the encoded state's circuit
            return circuit

        self.encoder = encode_frqi

    def encode_state(self,
                     data: data_type) -> Circuit:
        """ Encode a classical data into a quantum state.

        Paramaters
        ----------
        `data` (data_type):
            The data to be encoded.

        Returns
        -------
        `circuit` (Circuit): The encoded state's circuit.
        """
        return self.encoder(data)

    def decode_state(self,
                     data: data_type,
                     encoded_state: Circuit,
                     log=True) -> Data:
        """ Decode a quantum state into the original classical data.

        Parameters
        ----------
        `data` (data_type):
            The original data for denormalization purposes.
        `encoded_state` (Circuit):
            The encoded state's circuit.
        `log` (bool):
            Whether to print the log or not.

        Returns
        -------
        `decoded_state` (Data): The decoded data.
        """
        # Define a Data instance
        if not isinstance(data, Data):
            data = Data(data)

        # Pad the data (if required)
        if data.padded is False:
            data.pad()

        if log:
            print(f"The depth of the circuit is : {encoded_state.get_depth()}")

        # Start timer (Measuring the circuit)
        start = time.time()

        # Run circuit (due to the additional qubit, we must add the splice at the end)
        state_vector = encoded_state.get_statevector()[2**(encoded_state.num_qubits-1):]

        # Convert state vector to a Data instance to apply further processing
        state_vector = Data(state_vector)

        # End timer (Measuring the circuit)
        end = time.time()

        if log:
            print('Measurement done...')
            print(f'Time taken to measure : {end-start}')

        # Start timer (Decoding the state vector)
        start = time.time()

        # Denormalize the state vector and reshape the decoded state into the original shape
        decoded_state = Data.denormalize_data(state_vector.data, data.norm_scale, data.shape)

        # End timer (Decoding the state vector)
        end = time.time()

        if log:
            print('Decoding done...')
            print(f'Time taken to decode : {end-start}')

        # Return the decoded data
        return Data(decoded_state)
```
